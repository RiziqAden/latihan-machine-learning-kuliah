#SOAL

Soal: Gunakan dataset dari Scikit Learn yaitu Diabetes, kemudian ambil data diabetes dari 'data' sebagai variabel dependen dan data Body Mass Index (BMI) dari 'target' sebagai variabel independen. Lalu pisahkan data training 75% dan data test 25% dari total data

Gunakan Regression Model yaitu Linear Regression, kemudian Ridge dan Lasso Regression dengan lambda = 0.25



berikan hasil MSE dan visualisasikan garis regresi beserta data points dari data test!



*NOTE: Gunakan '.reshape(-1,1)' pada variabel independen sebelum dilakukan split training dan testing



# load data

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_diabetes
import numpy as np

data_diabetes = load_diabetes()
```

# melihat data

```python
print("Jenis isian dari dataset diabetes = \n{}".format(data_diabetes.keys()))
```

```python
print("melihat kolom 'data' = \n{}".format(data_diabetes['data']))
```

```python
print("melihat kolom 'target' = \n{}".format(data_diabetes['target']))
```

# membagi data

```python
from sklearn.model_selection import train_test_split
#  x = data, y =target
x = data_diabetes.data[:,2].reshape(-1, 1) # ambil 2 fitur aja, yaitu data dan targer
y = data_diabetes.target

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=1) # bagi data 25%
```

# LinearRegression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error #semakin kecil (mendekati 0) MSE, semakin bagus


linear_model = LinearRegression()
#training
linear_model.fit(x_train,y_train)

#testing
y_pred = linear_model.predict(x_test)
mse = mean_squared_error(y_pred,y_test) #hasil prediksi dan data asli
print(f"hasil MSE = {mse}")

plt.scatter(x_test, y_test, color='blue', label = 'data sebenarnya')
plt.plot(x_test,y_pred, color = 'red', label = 'nilai prediksi')
plt.legend()
plt.title(f'Linear Regression \n MSE = {mse}')
plt.show()
```

# Ridge

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error #semakin kecil (mendekati 0) MSE, semakin bagus


ridge_model = Ridge(alpha=0.25) #lambda = alpha untuk mengukur seberapa besar penaltinya
#training
ridge_model.fit(x_train,y_train)

#testing
y_pred = ridge_model.predict(x_test)
mse = mean_squared_error(y_pred,y_test) #hasil prediksi dan data asli
print(f"hasil MSE = {mse}")

plt.scatter(x_test, y_test, color='blue', label = 'data sebenarnya')
plt.plot(x_test,y_pred, color = 'red', label = 'nilai prediksi')
plt.legend()
plt.title(f'Ridge \n MSE = {mse}')
plt.show()
```

# Lasso Regression

```python
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_squared_error #semakin kecil (mendekati 0) MSE, semakin bagus


lasso_model = Lasso(alpha=0.25) #lambda = alpha untuk mengukur seberapa besar penaltinya
#training
lasso_model.fit(x_train,y_train)

#testing
y_pred = lasso_model.predict(x_test)
mse = mean_squared_error(y_pred,y_test) #hasil prediksi dan data asli
print(f"hasil MSE = {mse}")

plt.scatter(x_test, y_test, color='blue', label = 'data sebenarnya')
plt.plot(x_test,y_pred, color = 'red', label = 'nilai prediksi')
plt.legend()
plt.title(f'Lasso Regression \n MSE = {mse}')
plt.show()
```
