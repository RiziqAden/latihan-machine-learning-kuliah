## Begging (bootstrap aggregating)

Bagging merupakan teknik ensemble yang menggunakan beberapa model dasar, kemudian dilatih secara independen pada subset data ynga diambil sampling (sampling dengan penggantian). Prediksi akhir dilakukan dengan cara menggabungkan hasil semua model, biasanya dengan voting (untuk klasifikasi) atau rata-rata (untuk regresi)

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.35, random_state = 0)

#Random Forest Regressor
rf_model = RandomForestRegressor(n_estimators = 100, random_state = 0)
rf_model.fit(x_train, y_train)

#predik
y_pred = rf_model.predict(x_test)

# untuk cek Overfitting atau underfitting
y_pred_train = rf_model.predict(x_train)

#evaluasi
print(f'train MSE = {mean_squared_error(y_pred_train,y_train)}')
print(f'MSE = {mean_squared_error(y_pred,y_test)}')
print(f'train MAE = {mean_absolute_error(y_pred_train,y_train)}')
print(f'MAE = {mean_absolute_error(y_pred,y_test)}')
```

## Boosting

Boosting merupakan teknik ensemble yang menggabungkan model dasar secara berurutan. setiap model baru mencoba untuk memperbaiki kesalahan dari model sebelumnya dengan memberikan bobot lebih besar pada data yang sulit di prediksi

###Adaboost

Membangun Model secara bertahap, dimana model berikut berfokus pada kesalahan model sebelumnya

```python
from sklearn.ensemble import AdaBoostRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.35, random_state = 0)

#Base Model
base_model = DecisionTreeRegressor(max_depth = 10)

#AdaBoost Regressor
adaboost_model = AdaBoostRegressor(estimator = base_model, n_estimators=100, random_state=0)
adaboost_model.fit(x_train,y_train)

#Prediksi
y_pred = adaboost_model.predict(x_test)
y_pred_train = adaboost_model.predict(x_train)

# Evaluasi
print(f'MSE = {mean_squared_error(y_pred,y_test)}')
print(f'MSE = {mean_squared_error(y_pred_train,y_train)}')
print(f'MAE = {mean_absolute_error(y_pred,y_test)}')
print(f'MAE = {mean_absolute_error(y_pred_train,y_train)}')
```

### Gradient Boosting:

merupakan teknik ensemble dengan meminimalkan loss function dengan cara menyesuaikan model baru terhadap kesalahan

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.35, random_state = 0)

# Gradient Boosting Regressor
gb_model = GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, max_depth=7,random_state=0)
gb_model.fit(x_train, y_train)

# predik
y_pred = gb_model.predict(x_test)

# evaluasi
print(f'MSE = {mean_squared_error(y_pred,y_test)}')
print(f'MAE = {mean_absolute_error(y_pred,y_test)}')
```

### XGBoost:

Merupakan implementasi yang lebih cepat dan dioptimalkan dari Gradient Boosting, Kelebihannya adalah mendukung paralelisasi, regularisasi, dan dapat mengendalikan overfitting

```python
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.2, random_state = 0)

# XGB Regressor
xgb_model = xgb.XGBRegressor(n_estimators=100, learning_rate=0.1, max_depth=10,random_state=0)
xgb_model.fit(x_train, y_train)

# predik
y_pred = xgb_model.predict(x_test)

# evaluasi
print(f'MSE = {mean_squared_error(y_pred,y_test)}')
print(f'MAE = {mean_absolute_error(y_pred,y_test)}')
```

## Stacking

Staking adalah teknik ensemble yang menggabungkan prediksi dari beberapa model dasar dengan menggabungkan meta-model(model tingkat kedua) untuk membuat prediksi akhir.

```python
from sklearn.ensemble import StackingRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.34, random_state = 0)

#Base Model
base_model = [
    ('rf',RandomForestRegressor(n_estimators=100, random_state=0)),
    ('svr', SVR(kernel='linear'))
]
#Meta Learner
meta_learner = LinearRegression()

#Stacking Re
stacking_model = StackingRegressor(estimators = base_model, final_estimator = meta_learner)
stacking_model.fit(x_train,y_train)

#Prediksi
y_pred = stacking_model.predict(x_test)
# evaluasi
print(f'MSE = {mean_squared_error(y_pred,y_test)}')
print(f'MAE = {mean_absolute_error(y_pred,y_test)}')
```
