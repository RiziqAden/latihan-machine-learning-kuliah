# Linear Regression

```python
data.columns
```

```python

x = data[['Hours_Studied',	'Attendance',	'Sleep_Hours','Previous_Scores']]
y = data[['Exam_Score']]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error


x = data[['Hours_Studied',	'Attendance',	'Sleep_Hours','Previous_Scores']]
y = data[['Exam_Score']]


# split
x_train, x_test, y_train, y_test = train_test_split(x, y,
                                                    test_size = 0.2,
                                                    random_state = 0)

model = LinearRegression()
model.fit(x_train, y_train)

y_pred = model.predict(x_test)

mse = mean_squared_error(y_pred, y_test)
print(f'Hasil MSE = {mse:.3f}')
```

```python
model.coef_
```

# Feature Importance

```python
import numpy as np

feature_importance = pd.DataFrame({
    'Feature': x.columns,
    'Importance': model.coef_[0]
})

feature_importance = feature_importance.sort_values(by='Importance', ascending =False)
print(feature_importance)
```

Berdasrkan Hasil feature importance dari model Linear Regresion, dapat dilihat bahwa Hours_studied (jam belajar perminggu) memiliki nilai kepentingan (importance) yang paling tinggi.

sehingga menunjukkan pengaruh paling besar terhadap Exam Score (nilai ujian)

attendace(kehadiran) juga memberikan pengaruh terhadap Exam Score, Namun tidak lebih tinggi di bandingkan Hours_studied, begitu juga dengan Previous_Scores.

SEDANGKAN Sleep_hours Dikarenakan hasil berupa minus pada nilai importance, sehingga tidak memberikan pengaruh terhadap prediksi model atau terhadap Exam Scores

# Klasifikasi / classification

```python
# label Encoding
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
data['Parental_Involvement_encoded'] = label_encoder.fit_transform(data['Parental_Involvement'])

data
```

```python
label_mapping = dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_)))

print('Parental_Involvement_encoded: ',label_mapping)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score


x = data[['Hours_Studied',	'Attendance',	'Sleep_Hours','Previous_Scores']]
y = data[['Parental_Involvement_encoded']]


# split
x_train, x_test, y_train, y_test = train_test_split(x, y,
                                                    test_size = 0.2,
                                                    random_state = 0)

model = LogisticRegression()
model.fit(x_train, y_train)

y_pred = model.predict(x_test)

akurasi = accuracy_score(y_pred, y_test)
print(f'Hasil akurasi model = {akurasi:.2%}\n')

# Apakah Feature Encoded bisa digunakan sebagai Features ? Jawaban BISAA
```

```python
model.coef_[0]
```

```python
feature_importance = pd.DataFrame({
    'Feature': x.columns,
    'Importance': np.abs(model.coef_[0]) # KHUSUS KLASIFIKASI, PAKAI
})

feature_importance = feature_importance.sort_values(by='Importance', ascending =False)
print(feature_importance)
```

Importance bisa lebih dari 1 berarti bisa mempengaruh banget
