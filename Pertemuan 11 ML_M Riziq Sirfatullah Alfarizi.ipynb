# Linear Discriminant analysis (LDA)

Definisi: <br>
LDA merupakan teknik reduksi dimensi yang memanfaatkan label kelas untuk memaksimalkan separasi antar kelas.

Keunggulan:
- memaksimalkan separasi antar kelas
- cepat dalam komputasi dan cocok untuk data berukuran kecil
- sangat baik untuk klasifikasi dengan jumlah fitur yang besar

Dibandungkan PCA yang bisa digunkan untuk regresi maupun klasifikasi. LDA lebih cocok digunakan untuk data klasifikasi namun tidak cocok untuk regresi

Khusus untuk tujuan klasifikasi, LDA dapat membantu menaikkan akurasi dibandingkan PCA. jika penggunaan LDA memberikan hasil akurasi lebih rendah dibandingankan PCA, terdapat beberapa kondisi, yaitu:
- kelas tidak terbagi rata dengan baik
- dataset Non linear
- Terbapat Outlier/noise
-

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
iris = load_iris()
x = iris.data
y = iris.target # target -> label
```

stratify = y

merupakan parameter yang berfungsi untuk memastikan bahwa proporsi label kelas dalam data training dan data testing sama.

```python
x_train, x_test,  y_train, y_test = train_test_split(x, y, random_state=0, stratify=y)
```

Selanjutnya uji coba digunakan PCA dan LDA tanpa preprocessing StandarScaler

```python
# Random Forest Classifier
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(random_state=0)
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
akurasi = accuracy_score(y_test, y_pred)

print(f"Hasil Akurasi Menggunakan RandomForestClassifier = {akurasi:.2%}")
```

```python
# Random Forest Classifier menggunakan PCA
pca = PCA(n_components=2) # diubah menjadi 2 feature
x_tain_pca = pca.fit_transform(x_train)
x_test_pca = pca.transform(x_test)

model_pca = RandomForestClassifier(random_state=0)
model_pca.fit(x_tain_pca, y_train)
y_pred_pca = model_pca.predict(x_test_pca)
akurasi = accuracy_score(y_test, y_pred_pca)

print(f"Hasil Akurasi Menggunakan RandomForestClassifier dan PCA = {akurasi:.2%}")
```

```python
# Random Forest Classifier menggunakan LDA
# Syarat -> Jumlah Komponen < jumlah label
# Misal Jumlah label ada 3 -> berarti jumlah komponen harus 2 atau 1

lda = LDA(n_components=2) # diubah menjadi 2 feature
x_tain_lda = lda.fit_transform(x_train, y_train) # perubahan membutuhkan data x dan y
x_test_lda = lda.transform(x_test) # langsung transform saja

model_lda = RandomForestClassifier(random_state=0)
model_lda.fit(x_tain_lda, y_train)
y_pred_lda = model_lda.predict(x_test_lda)
akurasi = accuracy_score(y_test, y_pred_lda)

print(f"Hasil Akurasi Menggunakan RandomForestClassifier dan PCA = {akurasi:.2%}")
```
