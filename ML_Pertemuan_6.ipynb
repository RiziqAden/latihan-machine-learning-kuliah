# Soal Tugas

Lakukan web scraping pada link di bawah ini yaitu Real GDP Growth Rate dengan mengambil 150 baris pertama dari tahun 2020 sampai 2022 dengan target diambil tahun 2023, hitung MSE menggunakan seluruh metode regresi yang pernah dipelajar

# Web Scraping

Web Scraping Adalah proses otomatis untuk mengunduh/mengambil informasi dari halaman website. teknik ini menggunakan program untuk membaca konten website kemudian mengambil data yang diinginkan, seperti gambar, teks, tabel DLL dari format HTML

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

url = 'https://en.wikipedia.org/wiki/List_of_countries_by_real_GDP_growth_rate'
page = requests.get(url)

soup = BeautifulSoup(page.text,'html')
```

```python
table = soup.find_all('table', class_ = 'wikitable sortable static-row-numbers static-row-header-text')[2] # mengambil table ke 3
```

```python
table.find_all('th')
```

```python
judul_kolom = table.find_all('th')

list_judul_kolom = [judul.text for judul in judul_kolom]
print(list_judul_kolom)
```

```python
list_judul_kolom = [judul.text.strip() for judul in judul_kolom]
print(list_judul_kolom)
```

```python
df = pd.DataFrame(columns = list_judul_kolom)
df
```

```python
table.find_all('tr')
```

```python
data_kolom = table.find_all('tr')
```

```python
for baris in data_kolom:
  data_baris = baris.find_all('td')
  data_baris_individu = [data.text.strip() for data in data_baris]
  print(data_baris_individu)
```

# Mengambil 150 baris pertama dan Pindahkan data ke DataFrame

```python
for baris in data_kolom[1:151]: # baris di mulai dari 1 dan seterusnya
  data_baris = baris.find_all('td')
  data_baris_individu = [data.text.strip() for data in data_baris]
  print(data_baris_individu)
  # pindah data ke dataframe
  panjang = len(df)
  df.loc[panjang] = data_baris_individu
```

```python
df
```

# Cleaning Data

```python
import numpy as np

df.replace(['n/a','na'], np.nan, inplace=True)
```

```python
df['2013'][0]
```

```python
type(df['2013'][0])
```

```python
for kolom in df.columns[1:]: # karna index negara tidak dijadikan float
  df[kolom] = pd.to_numeric(df[kolom], errors= 'coerce').astype('float64')

df
```

```python
type(df['2013'][0])
```

```python
for kolom in df.columns[1:]:
  df[kolom].fillna(0, inplace=True)
  # df[kolom].fillna(df[kolom].mean(), inplace=True)

df
```

# train test split

```python
from sklearn.model_selection import train_test_split

kolom_x = ['2020','2021','2022']

x = np.array(df[kolom_x])
y = np.array(df['2023'])

x_train, x_test, y_train, y_test = train_test_split(
    x, y,
    test_size=0.2,
    random_state=0
)
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error #semakin kecil (mendekati 0) MSE, semakin bagus


linear_model = LinearRegression()
#training
linear_model.fit(x_train,y_train)

#testing
y_pred = linear_model.predict(x_test)
mse = mean_squared_error(y_pred,y_test) #hasil prediksi dan data asli
print(f"hasil MSE = {mse}")
```

# Ridge Regression

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error #semakin kecil (mendekati 0) MSE, semakin bagus


ridge_model = Ridge() #lambda = alpha untuk mengukur seberapa besar penaltinya
#training
ridge_model.fit(x_train,y_train)

#testing
y_pred = ridge_model.predict(x_test)
mse = mean_squared_error(y_pred,y_test) #hasil prediksi dan data asli
print(f"hasil MSE = {mse}")
```

# Lasso Regression

```python
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_squared_error #semakin kecil (mendekati 0) MSE, semakin bagus


lasso_model = Lasso() #lambda = alpha untuk mengukur seberapa besar penaltinya
#training
lasso_model.fit(x_train,y_train)

#testing
y_pred = lasso_model.predict(x_test)
mse = mean_squared_error(y_pred,y_test) #hasil prediksi dan data asli
print(f"hasil MSE = {mse}")
```

# PolyRegresi

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline

PolyRegresi = make_pipeline(PolynomialFeatures(degree=5),LinearRegression())
PolyRegresi.fit(x_train,y_train)

#testing
y_pred = PolyRegresi.predict(x_test)
mse = mean_squared_error(y_pred,y_test) #hasil prediksi dan data asli
print(f"hasil MSE = {mse}")

```
