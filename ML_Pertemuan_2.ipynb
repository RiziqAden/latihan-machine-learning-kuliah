# Klasifikasi

apa itu klasifikasi? <br>
klasifikasi adalah proses untuk mengidentifikasi dari model untuk menghasilkan data sebutan dari proses training

# Load data

```python
from sklearn.datasets import load_iris
iris_datasets = load_iris()
```

```python
print("Jenis isian dari dataset iris = \n{}".format(iris_datasets.keys()))
```

```python
print("Target name = \n{}".format(iris_datasets['target_names']))
```

```python
print("Feature name = ",format(iris_datasets['feature_names']))
```

```python
print("data name = \n{}".format(iris_datasets['data'][:5])) #menampilkan 5 data saja
```

```python
print("Target = \n{}".format(iris_datasets['target']))
# jenis bunga iris
# 0 = 'setosa', 2 = 'versicolor', 3 = 'virginica'
```

```python
# melihat tipe data pada data
print('tipe data = {}'.format(type(iris_datasets['data'])))
print('tipe target = {}'.format(type(iris_datasets['target'])))
```

```python
print('ukuran data, (kolom)= {}'.format(iris_datasets['data'].shape)) # 150 baris dan 4 kolom
print('ukuran target = {}'.format(iris_datasets['target'].shape)) # 150 baris dan 1 kolom
```

# proses Training dan Testing

langkah pertama: membagi data 150 data menjadi data untuk training dan data untuk testing <br>
kita coba membagi data Training (75%) dan data testing(25%)

```python
from sklearn.model_selection import train_test_split
#  x = data, y =target
x_train, x_test, y_train, y_test = train_test_split(
    iris_datasets['data'],iris_datasets['target'], test_size=0.25,random_state=0)
# rando_state = 0 -> kalau di run ulang , isi data yang dibagi tetap sama
```

```python
# jumlah data training
print('x_train shape = {}'.format(x_train.shape)) # data = (['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)'])
print('y_train shape = {}'.format(y_train.shape)) # Target = (sentosa, versicolor, viginica)
```

```python
# jumlah data training
print('x_test shape = {}'.format(x_test.shape)) # data = (['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)'])
print('y_test shape = {}'.format(y_test.shape)) # Target = (sentosa, versicolor, viginica)
```

# proses training

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(x_train,y_train)
```

melakukan presiksi

```python
import numpy as np
x_baru = np.array([[3, 2.1, 5, 1.2]])
prediksi = knn.predict(x_baru)
print('hasil prediksi Model knn = {}'.format(prediksi))
print('hasil prediksi Model knn pake nama = {}'.format(iris_datasets['target_names'][prediksi]))
```

# Evaluasi Model


```python
y_pred = knn.predict(x_test)
print(y_pred)
```

```python
print('hasil Akurasi model = {}'.format(knn.score(x_test, y_test)))
# hasil akurasi 97% bener
```

```python
from sklearn.datasets import load_breast_cancer
data_cancer = load_breast_cancer()
```

```python
print("Jenis isian dari dataset iris = \n{}".format(data_cancer.keys()))
```

```python
print("Target name = \n{}".format(data_cancer['target_names']))
```

```python
print("Feature name = ",format(data_cancer['feature_names']))
```

```python
print('ukuran data, (kolom)= {}'.format(data_cancer['data'].shape))
print('ukuran target = {}'.format(data_cancer['target'].shape))
```

```python
print("data name = \n{}".format(data_cancer['data'][:5])) #menampilkan 5 data saja
```

## training dan test

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(
    data_cancer['data'],data_cancer['target'], test_size=0.25,random_state=0)
```

```python
# jumlah data training
print('x_train shape = {}'.format(x_train.shape))
print('y_train shape = {}'.format(y_train.shape))
```

```python
from sklearn.neighbors import KNeighborsClassifier
model_cancer = KNeighborsClassifier(n_neighbors=1)
```

```python
model_cancer.fit(x_train,y_train)
```

## Evaluasi

```python
print('hasil Akurasi model = {}'.format(model_cancer.score(x_test, y_test)))
```
