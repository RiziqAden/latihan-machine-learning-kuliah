# <B> Principal Component Analysis</b>

Apa itu PCA?

PCA Merupakan metode statistik yang digunakan untuk <b> Mereduksi Dimensi </b> dengan cara mengubah data ke dalam komponen Utama (Orincipal Coponent) yang memiliki variansi terbesar. PCA digunakan untuk mengatasi masalah dimensi pada data yang memiliki banyak fitur, tanpa kehilangan terlalu banyak informasi penting

<b>Manfaat PCA: </b>

1. Reduksi Dimensi
2. Visualisasi data
3. Menguragi Overfitting (Menghapus Fitur kurang penting)
4. Mempercepat komputasi

# Menggukan PCA

```python
x = data[['Algebra', 'Calculus1',
       'Calculus2', 'Statistics', 'Probability', 'Measure',
       'Functional_analysis']]
y = data['y']
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Preprocessing
# Normalisasi data (Data di-scaling agar memiliki varians yang seragam)
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

```python
pca = PCA(n_components=2) # mengurangi feature menjadi 2 feature
x_pca = pca.fit_transform(x_scaled)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x_pca, y, test_size=0.2, random_state=0)
```

```python
# Support Vector Classidication (SVC)
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

model_svc2 = SVC()
model_svc2.fit(x_train, y_train)
y_pred = model_svc2.predict(x_test)
akurasi = accuracy_score(y_test, y_pred)

print(f"Hasil Akurasi Menggunakan SVM = {akurasi:.2%}")
```

Berdasarkan Hasil pertama, data yang digunkan untuk training hanya 2 feature saja memberikan akurasi sebesar 61.9%.

sedangkan data dengan 5 feature kemudian direduksi menjadi 2 fitur agar informasi data sebelumnya tidak hilang, memberikan hasil akurasi sebesar 66.7%

KESIMPULAN

Untuk dapat mempertahankan tanpa mengurangi akurasi yang besarm disarankan untuk mengambil fitur sebanyak-banyaknya dan melalukan reduksi menggunakan PCA jika tujuaanya untuk Visualisasi 2 dimensi (2 Feature) atau 3 dimensi (3 feature)

# Feature Importance & PCA

```python
data.info()
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
data['class_encode'] = label_encoder.fit_transform(data['class'])
data['gender_encode'] = label_encoder.fit_transform(data['gender'])
data['from1_encode'] = label_encoder.fit_transform(data['from1'])
data['from2_encode'] = label_encoder.fit_transform(data['from2'])
data['from3_encode'] = label_encoder.fit_transform(data['from3'])
```

```python
# menghapus data kategori
data2 = data.drop(columns=['ID','class', 'gender', 'from1', 'from2', 'from3'])
data2
```

```python
data2.columns
```

```python
x = data2.drop(columns=['y'])
y = data2['y']
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
# Support Vector Classidication (SVC)
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

model_svc = SVC(kernel = 'linear')
model_svc.fit(x_train, y_train)
y_pred = model_svc.predict(x_test)
akurasi = accuracy_score(y_test, y_pred)

print(f"Hasil Akurasi Menggunakan SVM = {akurasi:.2%}")
```

```python
import numpy as np
feature_importance = pd.DataFrame({
    'Feature': x.columns,
    'Importance': np.abs(model_svc.coef_[0])
})
feature_importance = feature_importance.sort_values(by='Importance', ascending=False)
feature_importance
```

Berdasarkan Feature importance, terdapat 8 fitur yang nilainya kurang dari 0.1, Sehingga fitur tersebut tidak memberikan dampak besar pada model yg digunakan, Jadi perlu dilakukan training ulang dengan <b>mereduksi </b>dimensi menjadi 6 fitur, karena memberikan nilai lebih dari 0.1

```python
x = data2.drop(columns=['y'])
y = data2['y']
```
